```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv("Search Analysis/Queries.csv")
```

```python
df['CTR'] = df['CTR'].astype(str)
df['CTR'] = df['CTR'].str.replace('%', '')
df['CTR'] = df['CTR'].astype(float)
df[(df['Impressions'] > 1000) & (df['Impressions'] < 2000)].sort_values(by='CTR', ascending=False).reset_index(drop=True)
```
