```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/leather-company-sales/AnalyticsDataLast3Months.csv', skiprows=9)
df.head()
```

```python
df['PageType'] = df['Landing page + query string'].str.split('-').str[-1]
df['fPageType'] = df['PageType'].astype(str).str[0]
df['fPageType'].value_counts()
```

```python
cat_df = df[df['fPageType'] == 'c']
prod_df = df[df['fPageType'] == 'p']
other_cat_df = df[(df['fPageType'] != 'p') & (df['fPageType'] != 'c')]
```

```python
cat_df.shape, prod_df.shape, other_cat_df.shape
```

```python
print('Main Category: ', cat_df['Organic Google Search impressions'].sum() / len(cat_df))
print('Product: ', prod_df['Organic Google Search impressions'].sum() / len(prod_df))
print('Other Categories: ', other_cat_df['Organic Google Search impressions'].sum() / len(other_cat_df))
```

```python
print('Main Category: ', cat_df['Organic Google Search clicks'].sum() / len(cat_df))
print('Product: ', prod_df['Organic Google Search clicks'].sum() / len(prod_df))
print('Other Categories: ', other_cat_df['Organic Google Search clicks'].sum() / len(other_cat_df))
```

```python
print('Main Category: ', cat_df['Engaged sessions'].sum() / len(cat_df))
print('Product: ', prod_df['Engaged sessions'].sum() / len(prod_df))
print('Other Categories: ', other_cat_df['Engaged sessions'].sum() / len(other_cat_df))
```

```python
print('Main Category: ', cat_df['Engagement rate'].sum() / len(cat_df))
print('Product: ', prod_df['Engagement rate'].sum() / len(prod_df))
print('Other Categories: ', other_cat_df['Engagement rate'].sum() / len(other_cat_df))
```
