```python
import pandas as pd
```

```python
df = pd.read_csv('Amazon-LC stock - Sheet1.csv')
```

```python
df.head()
```

```python
df['Product'] = df['Product'].astype(str)
```

```python
temp2_df = df['Product'].str.split('B0', expand=True)
temp2_df.columns = ['f_', 's_']
temp2_df['f_'] = temp2_df['f_'].str.replace('...', '')
temp2_df['s_'] = temp2_df['s_'].str.lstrip()
temp2_df['s_'] = temp2_df['s_'].str.rstrip()
temp2_df['s_'] = temp2_df['s_'].str.lstrip()
temp2_df['s_'] = temp2_df['s_'].str.rstrip()
df['ASIN'] = temp2_df['s_']
df['ASIN'] = 'B0' + df['ASIN']
temp2_df['s_'] = 'B0' + temp2_df['s_']
df['Product'] = temp2_df['f_']
```

```python
df = df[['ASIN', 'Product', 'Amazon stock', 'LC stock']]
df.to_csv('Amazon-LC stock with ASIN.csv')
```
