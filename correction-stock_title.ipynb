```python
df = pd.read_csv('vs_export_20240503-164449.csv', dtype=str)
print(df.shape)
orders_df = pd.read_csv("All orders - short custom description.csv", dtype=str)
orders_df = orders_df[['title', 'customs_description']]
# orders_df.rename(columns={'title': 'Parent Product Title'}, inplace=True)
orders_df = orders_df[orders_df['customs_description'] != 'Sale']
dict_customs_title = pd.Series(orders_df.customs_description.values,index=orders_df.title).to_dict()
df['customs_description'] = df['Parent Product Title']
df['customs_description'] = df['customs_description'].map(dict_customs_title)
temp = df[['customs_description', 'Parent Product Title', 'Categories']]
temp = temp[~temp.isnull().any(axis=1)]
dict_customs_category = pd.Series(temp.customs_description.values,index=temp.Categories).to_dict()
for i in temp_ladies_update:
    dict_customs_category.update({i: "Ladies > Leather Jackets"})
for i in temp_mens_update:
    dict_customs_category.update({i: "Mens > Leather Jackets"})


df['customs_description'] = df['Categories']
df['customs_description'] = df['customs_description'].map(dict_customs_category)
df['Sale Price (Inc VAT)'] = df['Sale Price (Inc VAT)'].astype(float)
df['Price (Inc VAT)'] = df['Price (Inc VAT)'].astype(float)
df['RRP Price (Inc VAT)'] = df['RRP Price (Inc VAT)'].astype(float)

df_rrp_zero = df[df['RRP Price (Inc VAT)'] == 0]
df_rrp_notzero = df[df['RRP Price (Inc VAT)'] != 0]
df_rrp_notzero['Price (Inc VAT)'] = df_rrp_notzero['RRP Price (Inc VAT)']
df = pd.concat([df_rrp_zero, df_rrp_notzero], ignore_index=True)
df['RRP Price (Inc VAT)'] = 0
df.shape
```
