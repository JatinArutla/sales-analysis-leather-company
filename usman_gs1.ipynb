```python
import pandas as pd
import numpy as np
```

```python
usman = pd.read_csv("All Usman Barcodes.csv", dtype=str)
usman['Product_Orig'] = usman['Product']
fashione = pd.read_csv('Fashione-EAN-PLU.csv', dtype=str)
fashione.dropna(subset=['EAN'], inplace=True)
fashione.rename(columns={'EAN': 'Number'}, inplace=True)
fashione = fashione[['Number', 'PLU']]
temp = pd.merge(usman, fashione, on='Number', how='left')
fashione1 = temp[~temp['PLU'].isnull()]
fashione1['On Fashione'] = 'Yes'
usman = temp[temp['PLU'].isnull()]
usman.drop(columns=['PLU'], axis=1, inplace=True)
print(fashione1.shape)
```

```python
df = pd.read_csv("Fashione-EAN-PLU.csv", dtype=str)
df['Size'] = df['Size'].replace('ONE SIZE', '')
df['Size'] = df['Size'].replace(np.NaN, '')
df['Style'] = df['Style'].str.lower()
df['Style'] = df['Style'].replace('-', '', regex=True)
df['Style'] = df['Style'].str.lstrip()
df['Style'] = df['Style'].str.rstrip()
df['Size'] = df['Size'].str.lstrip()
df['Size'] = df['Size'].str.rstrip()

df['Colour'] = df['Colour'].str.lower()
df['Colour'] = df['Colour'].str.replace('blk', 'black')
df['Colour'] = df['Colour'].str.lstrip()
df['Colour'] = df['Colour'].str.rstrip()
# df['Colour'] = df['Colour'].str.replace(np.NaN, '')
# temp_df = df['Colour'].str.split('/', expand=True)
# temp_df.columns = ['F_Colour', 'S_Colour', 'T_Colour', 'Fourth_Colour']
# df['Colour'] = temp_df['F_Colour']

df['Size'] = df['Size'].str.lower()
df['Product'] = df['Style'] + ' ' + df['Colour'] + ' ' + df['Size']

df = df[['PLU', 'Product']]
df.dropna(subset=['Product'], inplace=True)
df['Product'] = df['Product'].str.lstrip()
df['Product'] = df['Product'].str.rstrip()
```

```python
usman['Product'] = usman['Product'].str.lower()
usman['Product'] = usman['Product'].replace('-', '', regex=True)
```

```python
usman[usman['Product'].str.contains('g27')]
```

```python
temp = pd.merge(usman, df, on='Product', how='left')
temp = temp.sort_values(by=['PLU'], ascending=True)
print(temp.dropna(subset=['PLU']).shape)
temp['On Fashione'] = 'No'
```

```python
usman[usman['Product'].str.contains('g27')]
```

```python
usman = pd.concat([fashione1, temp], ignore_index=True)
usman['Product'] = usman['Product_Orig']
usman.drop(columns=['Product_Orig'], axis=1, inplace=True)
```

```python
usman.to_csv('Usman-Fashione-Link.csv', index=False)
```
